```python
from lxml import html
import requests
from bs4 import BeautifulSoup
import urllib

states_page = requests.get('http://berkeleyearth.lbl.gov/state-list/')
state_names = []

soup = BeautifulSoup(states_page.text, 'html.parser').find('div','main-content')
state_links = soup.find_all('a')

# Every other link is a country, which we don't want
for i, link in enumerate(state_links):
    if i % 2:
        state_links[i-1].string.replace_with(state_links[i-1].string + '::' + link.string)
state_links[::2]
```

```python
# Every other link is a country, which we don't want
for link in state_links[::2]:
    file_slug = urllib.parse.quote('-'.join(link.string.lower().split()).encode('latin-1'))
    state_name = link['href'].split('/')[-1]
    #print(state_name
    url = 'http://berkeleyearth.lbl.gov/auto/Regional/TAVG/Text/{}-TAVG-Trend.txt'.format(state_name)
    file_name = 'state2/{}-TAVG-Trend.txt'.format(file_slug)
    state_names.append(link.string)
    try:
        #urllib.request.urlretrieve(url, file_name)
        print('downloaded {} from {}'.format(file_name, url))
    except:
        print('could not download {} from {}'.format(file_name, url))
```
