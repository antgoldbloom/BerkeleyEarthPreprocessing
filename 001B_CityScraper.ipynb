```python
from lxml import html
import requests
from bs4 import BeautifulSoup
import urllib.request
import string

alphabet = list(string.ascii_uppercase)
cities_page = requests.get('http://berkeleyearth.lbl.gov/city-list')
#alphabetical_cities_pages = ['http://berkeleyearth.lbl.gov/city-list/' + x for x in alphabet]
#print(alphabetical_cities_pages)
cities_names = []

soup = BeautifulSoup(cities_page.text, 'html.parser').find('div','main-content').find('table')
cities_links = soup.find_all('a')
# Every second link is a country, which we do not need
cities_links[::2]
for i, link in enumerate(cities_links):
    if i % 2:
        cities_links[i-1].string.replace_with(cities_links[i-1].string + '::' + link.string)
#cities_links[::2]
```

```python
# Every other link is a country, which we don't want
for link in cities_links[::2]:
    coords = link['href'].split('/')[-1]
    file_slug = urllib.parse.quote('-'.join(link.string.lower().split()).encode('latin-1'))
    print(file_slug, coords)
    url = 'http://berkeleyearth.lbl.gov/auto/Local/TAVG/Text/{}-TAVG-Trend.txt'.format(coords)
    file_name = 'city_major/{}-TAVG-Trend.txt'.format(file_slug + '%3A%3A' + coords)
    cities_names.append(link.string)
    try:
        #urllib.request.urlretrieve(url, file_name)
        print('downloaded {} from {}'.format(file_name, url))
    except:
        print('could not download {} from {}'.format(file_name, url))
```

```python
for letter in alphabet:
    letter_cities_page = requests.get('http://berkeleyearth.lbl.gov/city-list/' + letter)

    soup = BeautifulSoup(letter_cities_page.text, 'html.parser').find('div','main-content').find('table')
    letter_cities_links = soup.find_all('a')
    # Every second link is a country, which we do not need
    for i, link in enumerate(letter_cities_links):
        if i % 2:
            letter_cities_links[i-1].string.replace_with(letter_cities_links[i-1].string + '::' + link.string)
    #print(letter_cities_links[::2])
    
    # Every other link is a country, which we don't want
    for link in letter_cities_links[::2]:
        coords = link['href'].split('/')[-1]
        file_slug = urllib.parse.quote('-'.join(link.string.lower().split()).encode('latin-1'))
        print(file_slug, coords)
        url = 'http://berkeleyearth.lbl.gov/auto/Local/TAVG/Text/{}-TAVG-Trend.txt'.format(coords)
        file_name = 'city2/{}-TAVG-Trend.txt'.format(file_slug + '%3A%3A' + coords)
        try:
            urllib.request.urlretrieve(url, file_name)
            print('downloaded {} from {}'.format(file_name, url))
        except:
            print('could not download {} from {}'.format(file_name, url))
```
