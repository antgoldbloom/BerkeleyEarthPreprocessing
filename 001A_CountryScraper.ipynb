```python
from lxml import html
import requests
from bs4 import BeautifulSoup
import urllib

countries_page = requests.get('http://berkeleyearth.lbl.gov/country-list/')
country_names = []

soup = BeautifulSoup(countries_page.text, 'html.parser').find('div','main-content')
country_links = soup.find_all('a')
country_links
```

```python
for link in country_links:
    file_slug = urllib.parse.quote('-'.join(link.string.lower().split()).encode('latin-1'))
    url = 'http://berkeleyearth.lbl.gov/auto/Regional/TAVG/Text/{}-TAVG-Trend.txt'.format(file_slug)
    file_name = 'country2/{}-TAVG-Trend.txt'.format(file_slug)
    country_names.append(link.string)
    try:
        urllib.request.urlretrieve(url, file_name)
        print('downloaded {} from {}'.format(file_name, url))
    except:
        print('could not download {} from {}'.format(file_name, url))
```
